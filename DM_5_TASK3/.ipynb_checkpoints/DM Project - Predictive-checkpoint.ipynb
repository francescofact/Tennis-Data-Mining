```python
#function to discretize the variables
#input: the dataset and the list of variables' names to discretize
def discretize_data(dataset, variables):
    for variable in variables:
        #get the unique variable's values
        var = sorted(dataset[variable].unique())
        
        #generate a mapping from the variable's values to the number representation  
        mapping = dict(zip(var, range(0, len(var) + 1)))

        #add a new colum with the number representation of the variable
        dataset[variable+'_num'] = dataset[variable].map(mapping).astype(int)
    return dataset
```

We want do discretize gender and hand attributes.

```python
#discretize the variables
variables = ['gender', 'hand']
train_set_discretized = discretize_data(train_set, variables)
test_set_discretized = discretize_data(test_set, variables)

train_set_discretized = train_set_discretized.drop(columns=['gender', 'hand'])
test_set_discretized = test_set_discretized.drop(columns=['gender', 'hand'])


train_set_discretized.rename(columns={"gender_num": "gender", "hand_num": "hand"}, inplace=True)
test_set_discretized.rename(columns={"gender_num": "gender", "hand_num": "hand"}, inplace=True)
```

## Model evaluation metrics

Reminder about model evaluation metrics:

- Recall: Predicted true positives / actual positives 
- Precision: Predicted true positives / predicted positives

The recall measure the percentage of true positives that our model predict w.r.t. the real number of positives that we have in the data set.

The precision measure how much (in percentage) of the positives returned by the model is actually a true positive.

# KNN

k-Nearest Neighbors (k-NN) is a conceptually and computationally simple classification method: for each new, unlabeled point, we calculate, for some integer k, its k nearest neighbors in the n-dimensional predictor variable space.
Then we look at the labels of these k neighbors (i.e. 'high rank' or 'low rank') and assign the label with the most hits to the new point
(e.g. if k = 5, 3 neighbors vote 'high rank' and 2 vote 'low', then our model labels the new point as 'high')

#### Choose optimal value for K

There are no pre-defined statistical methods to find the most favorable value of K.
In the case of a small number of neighbors k, the noise will have a higher influence on the result, and a large number of neighbors make it computationally expensive. 

The optimal K value usually found is the square root of N, where N is the total number of samples.
But we will use two plots to understand the optimal value.

```python
import math
k = math.sqrt(test_set_discretized.size)
k = math.floor(k)
```

```python
if (k % 2) == 0:
   k = k-1
```

```python
from sklearn import metrics

error_rate = []
acc = []

for i in range(1,k):
     knn = KNeighborsClassifier(n_neighbors=i)
     knn.fit(train_set_discretized,train_label)
     pred_i = knn.predict(test_set_discretized)
     acc.append(metrics.accuracy_score(test_label, pred_i))
     error_rate.append(np.mean(pred_i != test_label))

plt.figure(figsize=(15,6))
plt.plot(range(1,k),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=5)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error: ",min(error_rate),"at K =", error_rate.index(min(error_rate)))

plt.figure(figsize=(15,6))
plt.plot(range(1,k),acc,color = 'blue',linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=5)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy: ",max(acc),"at K =", acc.index(max(acc)))

k = error_rate.index(min(error_rate))
```

#### Knn with the optimal value for K

```python
knn = KNeighborsClassifier(n_neighbors= k, algorithm='kd_tree').fit(train_set_discretized, train_label)
```

```python
#prediction on the test test
test_pred_knn = knn.predict(test_set_discretized)
train_pred_knn = knn.predict(train_set_discretized)
```

# Rule based classifier

Classify records by using a collection of if...then...  rules.
A rule *r* covers an instance *x* if the attributes of the instance satisfy the condition of the rule.

The fraction of records that satisfy the condition of a rule is called rule coverage.
The fraction od record which satisfy the condition and also the class prediction is called accuracy.

A classifier can contain MUTUALLY EXCLUSIVE rules, but this is not mandatory:
a record can trigger more than one rule: how to deal with that?
- ordered rule set (follow the order)
    - rule-based ordering: rules ranked individually
    - class-based ordering: rules that belongs to the same class appear together
- apply some voting schema

In the case in which a rule based classfier is not exhaustive (not couting for all possible combination of attribute values) that a record does not match any of the rule. In this case we can exploit a default class.



### How to build the set of rules

#### Direct method

Extract rule from data (i.e. RIPPER, CN2..)

- Start from an empty rule
- For each class
    - Grow a rule using Learn-One-Rule function
    - remove training reecords covered by the rule
    - repeat until stop criterion is met
    
Rule growing is computationally expensive: use greedy approach: 
- General-to-specific: from empty rule build the specific rules
- Specific-to-general: from records (very specific rules) build more general rule

#### Indirect method

Extract rules from other classification models (e.g. decision tree..)

## RIPPER

- For 2-class problem, choose one of the classes as positive class, and the other as negative class
    - learn rules for positive class
    - negative class will be the default one
- For multi-class problem
    - Order the classes according to increasing class prevalence (fraction of instances that belong to a particular class)
    - Learn the rule set for smallest class first, treat the rest as negative class
    - Repeat with next smallest class as positive class


- Growing a rule:
     - Start from empty rule
     - Add conjuncts as long as they improve FOILs information gain
     - Stop when rule no longer covers negative examples
     - Prune the rule immediately: delete any final sequence of conditions that maximize  v = (p-n)/(p+n) where p is positive examples covered by the rule and n the negative ones.
     
- Each time a rule is added to the rule set, compute the new description length
    - Stop adding new rules when the new description length is *d* its longer than the smallest description length obtained so far

# Naive Bayes Classifier

Relationship between attributes and class lables may not be deterministic.
Bayesian Classifier exploit the **Bayes Theorem** that combines prior knowledge on the class labels with knowledge derivable from data.

**Bayes theorem** P(Y|X) = P(X|Y) P(Y) / P(X)

Learn the posterior P(Y|X) for every combination of X and Y such that a record X'can be classified by finding the class Y' that maximizes the posterior probability.
This is equivalent of choosing the value Y' that maximizes P(X'|Y')P(Y'). Now we want to estimate P(X'|Y').

In practice the Naive Bayes classifier estimates the likelihood by assuming that the attributes are conditionally indipendent given the class label Y'.
If all the attributes are conditionally indipendet given Y' we can avoid to compute the likelihood for every combination of X, we only need to estimate the likelihood for each X' given Y.

Thus, for each class Y the naive bayes classifier just compute the posterior by multiplying the prior to the product for all the X of P(X_i | Y).

This makes the computation easier, but we still need to compute P(X_i | Y).

This distribution is estimated by observing the data.
For example for categorical attributes: P(X = x | Y = y) = Nxy / Ny, where Nxy is the records with value x and outcome y and Ny is the total number of record with outcome y.

For continuos attributes we can:
- Discretize range into bins and then count records with class y falling in the range x_i (simulating categorical attributes)
- Assume attribute follows a normal distribution and use data to estimate distribution parameters (e.g. mean and standard deviation)

Apply the Naive Bayes model from sklearn

# Random Forest

```python
#to find the best set of parameter setting, we can run a grid search
from sklearn.ensemble import RandomForestClassifier

#define the parameters' values you want to try
param_dist = {"max_depth": [2,3,5,6,7,10,12,None],
              "max_features": sp_randint(1, len(train_set_discretized.iloc[0]) + 1),
              "min_samples_split": sp_randint(10, 51),
              "min_samples_leaf": sp_randint(10, 51),
              "bootstrap": [True, False],
              "criterion": ["entropy", "gini"],
              "class_weight":['balanced', None, {0:0.56,1:0.44}]} #0.56 and 0.44 is our data balance
#define the number of iters
n_iter_search = 50
#define the model
clf = RandomForestClassifier(n_estimators=30)
#define the grid search
grid_search = RandomizedSearchCV(clf, param_distributions=param_dist, 
                            n_iter=n_iter_search, 
                            n_jobs=10, 
                            scoring=make_scorer(accuracy_score))
#run the grid search
grid_search.fit(train_set_discretized, train_label)
grid_search_result = grid_search.cv_results_['params'][0]
```

```python
print('Best setting parameters ', grid_search_result)
print('Mean and std of this setting ', grid_search.cv_results_['mean_test_score'][0], 
      grid_search.cv_results_['std_test_score'][0])
```

```python
rf = RandomForestClassifier(n_estimators=30, 
                             criterion=grid_search_result['criterion'],
                             max_features=grid_search_result['max_features'],
                             max_depth=grid_search_result['max_depth'], 
                             min_samples_split=grid_search_result['min_samples_split'],
                             min_samples_leaf=grid_search_result['min_samples_leaf'],
                             bootstrap=grid_search_result['bootstrap'],
                             class_weight=grid_search_result['class_weight']) 
rf = rf.fit(train_set_discretized, train_label)
```

```python
#prediction on the test test
test_pred_rf = rf.predict(test_set_discretized)
train_pred_rf = rf.predict(train_set_discretized)
```

```python
#compute the performance of the model
report_scores(train_label, train_pred_rf)
report_scores(test_label, test_pred_rf)
```

# AUC plot

The Area Under the Curve (AUC) is the measure of the ability of a classifier to distinguish between classes and it can used as a summary of the ROC curve.
